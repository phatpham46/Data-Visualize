```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# <strong> Read data</strong>

```python
# Read the data
df = pd.read_csv('../data/final_data.csv')
df.head(10)
```

# <strong> Analytics for data visualization</strong>
### &#9889; **Question 1. Did the total number of new infections worldwide increase or decrease day by day in the last 1 week?**
There are following steps to solve this problem:
- Get the latest 7 days data to calculate a 1-week period. Then take the total number of infections by day in the whole world
- Finally use the line plot to visualize the number of new cases during the week

#### **Step 1. Get the latest 7 days data to calculate a 1-week period. Then take the total number of infections by day in the whole world**

```python
#Lấy dữ liệu từ ngày 4/3/2023 tới ngày 10/3/2023
df = df.loc[(df['Date'] >= '2023-03-04') & (df['Date'] <= '2023-03-10')]
df.head(10)
```

```python
#Tính tổng số ca nhiễm của từng ngày
df_allWorld = df.groupby('Date')['NewCases'].sum().reset_index()
```

```python
#### **Step 2. Finally use the line plot to visualize the number of new cases during the week.**
```

```python
plt.figure(figsize=(10,5))
sns.set_style("darkgrid")
sns.lineplot(x="Date", y="NewCases", data=df_allWorld)
plt.title('Total number of new infections worldwide in the last 1 week')
plt.xlabel('Day')
plt.ylabel('Total number of new infections')
plt.show()
```

&#9889; <font color="yellow">**Comment on the results.**</font>
- On the 9th day, the number of infections spiked rapidly, reaching the highest level of the week.
- But with day 5, the number of cases is quite low, only a small part of day 9. This shows that the number of infections spiked on day 9 due to a number of countries where the number of infections spiked very quickly.
- The total number of new cases for the week is almost 1 million. Proving one thing that the current covid 19 has not been thoroughly controlled. Most people still believe that it is necessary to live with the epidemic in the near future.

> Here we only consider all the countries with the number of new infections, the countries with the new deaths will be added

```python
#Get the top 10 countries with the number of new infections on March 9, 2023
df_09_03 = df.loc[df['Date'] == '2023-03-09']
df_09_03 = df_09_03.sort_values(by='NewCases', ascending=False)
df_09_03 = df_09_03[['Country', 'NewCases']]
df_09_03 = df_09_03.head(10)
df_09_03
```

```python
#Draw a bar chart showing the top 10 countries with the number of new infections on March 9, 2023
sns.set_style("darkgrid")
sns.barplot(x="Country", y="NewCases", data=df_09_03)
plt.title('Top 10 countries with the number of new infections on March 9, 2023')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of new infections')
plt.show()
```

&#9889; <font color="yellow">**Comment on the results.**</font>
- As seen, the US is the country with the highest number of infections on March 9. Most of the countries with the highest number of infections are developed and densely populated countries
- In just one day, the number of cases was recorded much more than other days. This is a very worrying sign of the epidemic situation in the US. Or it is possible that the number of cases is aggregated after 3 to 4 days, causing a huge increase in a few days.
