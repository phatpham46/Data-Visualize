```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import branca.colormap as cm
```

# <strong> Read data</strong>

```python
covid_df = pd.read_csv('../data/final_data.csv')
covid_df.head()
```

# <strong> Analytics for data visualization</strong>
### &#9889; **Question 6: The total number of cases of the country?( Draw world map)**
There are many steps to answer this question:
- Get the latest date data and get the total number of cases of the country
- Draw the world map by folium

#### **Step 1. Get the latest date data and get the total number of cases of the country.**

```python
#Get data on 10/3/2023
covid_df["Date"] = pd.to_datetime(covid_df["Date"])
covid_df = covid_df[covid_df["Date"] == "2023-03-10"]
covid_df.head(10)
```

#### **Step 2. Draw the world map by folium.**

```python
color_map = cm.linear.RdYlBu_06.scale(0, 1)
m = folium.Map(location=[30, 0], zoom_start=2)
for i, row in covid_df.iterrows():

    cases = row['TotalCases']
    color_value = cases / 1000000
    color = color_map(color_value)

    folium.Marker(
        location=[row["lat"],row["lon"]],
        radius=5,
        color=None,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        tooltip=row['Country'] + ' - Total case: ' + str(cases),
    ).add_to(m)

color_map.caption = 'Total Cases'
m.add_child(color_map)
m
```
