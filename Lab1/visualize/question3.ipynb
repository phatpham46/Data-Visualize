# <strong> Analytics for data visualization</strong>
### &#9889; **Question 3. Is there a link between the number of tests and the number of cases?**
There are following steps to solve this problem:
- Get the data to prepare visualization(Total number of tests and total number of cases)
- Create a scatter plot with number of tests on the x-axis and number of infections on the y-axis


#### **Step 1. Get the data to prepare visualization.**

```python
# Get data from an existing dataframe
data = df[['TotalCases', 'TotalTests']]
data.head(10)
```

#### **Step 2. Create a scatter plot with number of tests on the x-axis and number of infections on the y-axis.**

```python
# Draw a scatter plot with TotalTests x-axis, TotalCases y-axis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="TotalTests", y="TotalCases")

# Add regression plot
sns.regplot(data=data, x="TotalTests", y="TotalCases", scatter=False, color="red")

# Name the x and y axes, and the chart title
plt.xlabel("Total Tests")
plt.ylabel("Total Cases")
plt.title("Relation between Total Tests and Total Cases")

plt.show()
```

&#9889; <font color="yellow">**Comment on the results.**</font>
- From the graph, we can see that as the number of tests increases, so does the number of infections.
- However, when the number of tests increased to a certain extent, the number of infections did not increase anymore. This may be because the number of tests has reached a level sufficient to detect new infections.
- So this shows that testing is one of the important factors to detect new infections and control the disease in the most optimal way.
