```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Make a request to the website URL
url = 'https://www.worldometers.info/coronavirus/'
response = requests.get(url)
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Find the elements you want to extract
today = soup.find('table', {'id': 'main_table_countries_today'})
today_body = today.find('tbody')
table_header = today.find('thead')
rows1 = today_body.find_all('tr')

yesterday = soup.find('table', {'id': 'main_table_countries_yesterday'})
yesterday_body = yesterday.find('tbody')
rows2 = yesterday_body.find_all('tr')

yesterday2 = soup.find('table', {'id': 'main_table_countries_yesterday2'})
yesterday2_body = yesterday2.find('tbody')
rows3 = yesterday2_body.find_all('tr')
```

```python
header = [ele.text.strip() for ele in table_header.find_all('th')]
```

```python
today_data = []
for row in rows1:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    today_data.append(cols)

yesterday_data = []
for row in rows2:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    yesterday_data.append(cols)

yesterday2_data = []
for row in rows3:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    yesterday2_data.append(cols)
    
```

```python
df1 = pd.DataFrame(today_data, columns=header)
df1.drop("#", inplace=True, axis=1)

df2 = pd.DataFrame(yesterday_data, columns=header)
df2.drop("#", inplace=True, axis=1)

df3 = pd.DataFrame(yesterday2_data, columns=header)
df3.drop("#", inplace=True, axis=1)
df3
```

```python
df1.to_csv('../data/covid_3_10_2023.csv', index=False)
df2.to_csv('../data/covid_3_9_2023.csv', index=False)
df3.to_csv('../data/covid_3_8_2023.csv', index=False)
```
